### Save csv files loader

```python
def safe_read_csv(path: str, nrows: int | None = None) -> pd.DataFrame | None:
#  Safely reads a CSV. Returns None if the file does not exist or cannot be read.
    if not os.path.exists(path):
        print(f"[safe_read_csv] File does not exist: {path}")
        return None
    
    try:
        return pd.read_csv(path, low_memory=False, nrows=nrows)
    except Exception as exc:
        print(f"[safe_read_csv] Error reading {path}: {exc}")
        return None
```
