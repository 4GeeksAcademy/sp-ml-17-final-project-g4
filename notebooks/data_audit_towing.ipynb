### Numerical - Categorical conversion

```python
from pathlib import Path
import pandas as pd

RAW = Path("data/raw")
CSV = df

# 1) Quick schema (cargamos pocas filas para inspección)
df_head = pd.read_csv(CSV, nrows=5000)
print("Columns:", list(df_head.columns))
print("Rows in sample:", len(df_head))

# 2) Definir un subconjunto de columnas de interés (clasificación)
USECOLS = [
    # ID (lo ignoraremos al entrenar)
    "Service ID",

    # Potenciales features (categóricas razonables y numéricas útiles)
    "Vehicle Type", "Make and Model", "Service Type",
    "Payment Method", "Warranty Information", "Service Status",
    "Customer Type",

    "Service Duration Hours", "Estimated Cost", "Mileage at Service",
    "Technician Rating", "Urgency Level",  # si es numérica u ordinal
    # variables de grúa (SOLO para crear la etiqueta; luego se excluyen)
    "Towing Date", "Tow Distance Miles",
]

df_sample = pd.read_csv(CSV, usecols=USECOLS, nrows=200_000)  # muestreo rápido

# 3) Crear etiqueta (towing_required) — NO la uses como feature
df_sample["towing_required"] = (
    df_sample["Towing Date"].notna()
    | (df_sample["Tow Distance Miles"].fillna(0) > 0)
).astype(int)

print("Class balance (sample):")
print(df_sample["towing_required"].value_counts(normalize=True))

# 4) Guardar muestra procesada para EDA y baseline
OUT = Path("data/processed")
OUT.mkdir(parents=True, exist_ok=True)
df_sample.to_csv(OUT / "synth_sample_for_eda.csv", index=False)
print("Saved:", OUT / "synth_sample_for_eda.csv")
```
